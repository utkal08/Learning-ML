```python
import matplotlib.pyplot as plt
import seaborn as sns
    
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips.head()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time')
```

```python
sns.lineplot(data=tips,x='total_bill',y='tip',hue='time')
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic
```

```python
sns.kdeplot(data=tips,x='total_bill')
```

```python
from sklearn import preprocessing 
label_encoder  = preprocessing.LabelEncoder()
titanic['alive']=label_encoder.fit_transform(titanic['alive'])
    
```

```python
titanic
```

```python
sns.kdeplot(data=titanic,x='alive',hue='sex')
```
